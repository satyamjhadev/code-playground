```python
#pattern questions
f=int(input())
n=(f+1)//2
n1=(f)-n
i=1
while i<=n:
    space=1
    while space<=n-i:
        print(" ",end="")
        space=space+1
    j=1
    while j<=(2*i)-1:
        print("*",end="")
        j=j+1
    print()
    i=i+1
i=n1
while i>=1:
    space=1
    while space<=(n1-i+1):
        print(" ",end="")
        space=space+1
    j=1
    while j<=(2*i)-1:
        print("*",end="")
        j=j+1
    print()
    i=i-1

```

```python
n = int(input())
if n%2==0:
    x=int(n/2+1)
else:
    x=int(n/2+1) 
for i in range(x):
    for k in range(1,x-i):
        print (" ",end="")
    for j in range(0,2*(i+1)-1):
        print("*", end="")
    print()
x=n-x
for i in range(x,0,-1):
    for k in range(x-i+1,0,-1):
        print (" ",end="")
    for j in range(2*(i)-1,0,-1):
        print("*", end="")
    print()
```

```python
f=int(input())  # f=5
n=(f+1)//2     #n=3
n1=(f)//2     # n1=2
i=1
while i<=n:
    space=1
    while space<=i-1:
        print(" ",end="")
        space=space+1
    j=1
    while j<=i:
        print("*",end="")
        j=j+1
    print()
    i=i+1
i=1          #i=2
while i<=n1:
    space=1
    while space<=(n1-i):
        print(" ",end="")
        space=space+1
    j=1
    while j<=n1-i+1:
        print("*",end="")
        j=j+1
    print()
    i=i+1
```
